```python
print('test')

import pandas as pd
import numpy as np
import seaborn as sns
# %pip install plotl
import plotly.express as px
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("C:\\Users\\hurai\\Extras\\python-ka-chilla-2024\\00_datasets\\diamonds.csv")
df.head()
```

```python
df=df.sample(frac=0.1, random_state=42)  # Sample 10% of the data for faster plotting
print(df.shape)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
fig=px.scatter(df, x='carat', y='price', color='cut',size='depth',
               hover_data=['clarity', 'x', 'y', 'z'],
               title='Diamond Price vs Carat Weight',
               labels={'carat': 'Carat Weight', 'price': 'Price (USD)'},)
fig.show()
```

```python
# take mean of carat prices and line plot using plotly
mean_prices = df.groupby('carat')['price'].mean().reset_index()
fig = px.line(mean_prices, x='carat', y='price', title='Mean Price of Diamonds by Carat Weight',
              labels={'carat': 'Carat Weight', 'price': 'Mean Price (USD)'})
fig.show()
```

```python
#bar plot of entire data

fig = px.bar(df, x='cut', y='price', color='clarity',
             title='Average Price of Diamonds by Cut and Clarity')
fig.show()    
```

```python
# for each cut find means and std of price
cut_stats = df.groupby('cut')['price'].agg(['mean', 'std']).reset_index()
fig = px.bar(cut_stats, x='cut', y='mean', error_y='std',
             color='cut',
             labels={'mean': 'Average Price', 'std': 'Standard Deviation'},
             title='Average Price of Diamonds by Cut with Standard Deviation', )
fig.show()
```

```python
fig=px.scatter(df, x='carat', y='price', color='cut', facet_col='clarity',
                size='price',
               animation_frame='cut',
                hover_data=['x', 'y', 'z'])
fig.show()
```
